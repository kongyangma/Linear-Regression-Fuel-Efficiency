```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
columnname=['mpg','cylinders','displacement','horsepower','weight','acceleration','model_year','origin','car name']
cars=pd.read_csv('C:/Users/kongy/Documents/Python Scripts/Datasets/auto-mpg.data.txt',delim_whitespace=True,names=columnname)
cars.head(10)
```

```python
cars=cars.drop('car name',axis=1)
cars.head(10)
```

```python
#What we need to do is to look at the relationship between different data and mpg fuel efficiency
#First of all, we look at relationship between each characteristic and mpg
cars.describe()
```

```python
cars.info()
```

```python
cars=cars[cars.horsepower!='?']
```

```python
#Draw a scatterplot, show the relationship between each feature and mpg
fig=plt.figure(figsize=(8,20))
ax1=fig.add_subplot(5,1,1)
ax2=fig.add_subplot(5,1,2)
ax3=fig.add_subplot(5,1,3)
ax4=fig.add_subplot(5,1,4)
ax5=fig.add_subplot(5,1,5)
ax1.scatter(cars.cylinders,cars.mpg,label='a')
ax1.set_title('Cylinders')
ax2.scatter(cars.displacement,cars.mpg)
ax2.set_title('displacement')
ax3.scatter(cars.horsepower,cars.mpg)
ax3.set_title('horsepower')
ax4.scatter(cars.weight,cars.mpg)
ax4.set_title('weight')
ax5.scatter(cars.acceleration,cars.mpg)
ax5.set_title('acceleration')
plt.show()
```

```python
#Build model, relationship between weight and mpg
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(cars[['weight']],cars[['mpg']])
cars['predictions']=lr_model.predict(cars[['weight']])
cars.head(10)
```

```python
#Set color of scatter points
T = np.arctan2(cars['weight'],cars['mpg'])
plt.scatter(cars['weight'], cars['mpg'], c = T)
#alpha is to set transparency
plt.scatter(cars['weight'], cars['predictions'], c = 'grey', alpha = 0.5)
plt.show()
```

```python
#Find out parameters beta=cov(x,y)/var(x)
var=cars['weight'].var()
cov=cars[['weight','mpg']].cov().iloc[0][1]
beta=cov/var
beta
```

```python
#Now we calculate alpha=ybar-xbar*beta
alpha=cars['mpg'].mean()-beta*cars['weight'].mean()
alpha
#Conclusion:the linear equation is y=46.22-0.008x
```

```python
#Now start model assessment
#Calculate use r2_score function from sklearn.metrics
from sklearn.metrics import r2_score
r2_score = r2_score(cars['mpg'],cars['predictions'])

#Calculate use score function from LinearRegression
score = lr_model.score(cars[['weight']],cars['mpg'])
r2_score
#The value is 0.69, which means more than half of the data could be explained using linear regression model, but not so good
```

```python
#Calculate MSE
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(cars['mpg'], cars['predictions'])
mse
```

```python
#Build model with 3 variables, which are weight, horsepower, displacement
mul_lr_model=LinearRegression()
mul_lr_model.fit(cars[['weight','horsepower','displacement']],cars['mpg'])
cars['mul_predictions']=mul_lr_model.predict(cars[['weight','horsepower','displacement']])
cars.head(5)
```

```python
#Let's see if the score gets better with 3 variables
mul_score=mul_lr_model.score(cars[['weight','horsepower','displacement']],cars['mpg'])
mul_score
#better result
```
